```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=sns.load_dataset("titanic")
```

### Visualization on the basis of two values
- Categorical data
- Numerical data

#### Basic Graph Types
- Line Graph :shows trends over time
- Bar Graph : compare values between categories (discrete data)
- Pie Chart : compare values between categories (percentage data)
- Histogram : show distribution of numerical data
- Box Plot : statistical distribution of numerical data
- Heat Map : show matrix of values using color
- Scatter Plot : show relationship between two variables
- Bubble Chart : show relationship between three variable

```python
sns.countplot(df,x="class")
plt.title("passengers count by class")
plt.xlabel("class")
plt.ylabel("count")
plt.show()
```

```python
#Countplot : Survival based on gender
sns.countplot(df, x="survived", hue="sex",palette="RdBu")
plt.title("Survivied Vs Not Survived")
plt.show()
```

```python
sns.barplot(df, x="sex", y="fare",hue="class",)
plt.title("Average Fare by Passenger Class and Gender")

```

```python
sns.pointplot(df, x="class", y="survived",hue="sex", markers=["o","x"])
plt.title("Survival rate by Class and Gender")
plt.show()
```
